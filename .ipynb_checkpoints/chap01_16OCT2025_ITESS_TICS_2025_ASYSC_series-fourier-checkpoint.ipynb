```python
import numpy as np
import matplotlib.pyplot as plt
from sympy import fourier_series, pi, cos, sin, symbols
from IPython.display import display, Math

%matplotlib inline
```

```python
display(Math(r'''
f(x) = \frac{a_0}{2} + \sum_{n=1}^{\infty} \left[ a_n \cos\left(\frac{2\pi n x}{T}\right) + b_n \sin\left(\frac{2\pi n x}{T}\right) \right]
'''))

display(Math(r'''
a_0 = \frac{2}{T} \int_{-T/2}^{T/2} f(x)  dx, \quad
a_n = \frac{2}{T} \int_{-T/2}^{T/2} f(x) \cos\left(\frac{2\pi n x}{T}\right) dx, \quad
b_n = \frac{2}{T} \int_{-T/2}^{T/2} f(x) \sin\left(\frac{2\pi n x}{T}\right) dx
'''))
```

```python
def square_wave_fourier(x, n_terms):
    """
    Aproximación de una onda cuadrada usando los primeros n términos de su serie de Fourier.
    f(x) = \sum_{k=1}^{n} \frac{\sin((2k-1)x)}{2k-1}
    """
    result = np.zeros_like(x)
    for k in range(1, n_terms + 1):
        n = 2 * k - 1  # solo términos impares
        result += np.sin(n * x) / n
    return (4 / np.pi) * result

# Parámetros
x = np.linspace(-2*np.pi, 2*np.pi, 1000)
n_terms = 10  # Número de términos en la expansión
```

```python
plt.figure(figsize=(10, 6))
plt.plot(x, square_wave_fourier(x, n_terms), label=f'Serie de Fourier ({n_terms} términos)', color='red')
plt.axhline(1, color='k', linestyle='--', alpha=0.5)
plt.axhline(-1, color='k', linestyle='--', alpha=0.5)
plt.xlim(-2*np.pi, 2*np.pi)
plt.ylim(-2, 2)
plt.title('Aproximación de una Onda Cuadrada con Serie de Fourier')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
from sympy import Piecewise, symbols, pi, sin, cos, Function, lambdify
from sympy.series.fourier import fourier_series
from IPython.display import display, Math
import numpy as np
import matplotlib.pyplot as plt  # Importar matplotlib para graficar

x = symbols('x')

# Definir la onda cuadrada con Piecewise
f = Piecewise(
    (1, (x > 0) & (x < pi)),   # f(x) = 1 en (0, π)
    (-1, (x > -pi) & (x < 0))  # f(x) = -1 en (-π, 0)
)

# Mostrar la definición de la función
display(Math(r'\text{Onda cuadrada definida como: }'))
display(f)

# Calcular la serie de Fourier
fs = fourier_series(f, (x, -pi, pi))

# Mostrar los primeros 5 términos
display(Math(r'\text{Expansión en serie de Fourier (primeros 5 términos):}'))
# CORRECCIÓN: Usar sympy.latex() para convertir la expresión a LaTeX
from sympy import latex
display(Math(latex(fs.truncate(5))))

# Convertir la serie truncada a una función numérica
f_approx = fs.truncate(10)
f_lambdified = lambdify(x, f_approx, 'numpy')

# Generar datos para graficar
x_vals = np.linspace(-2*np.pi, 2*np.pi, 1000)
y_vals = f_lambdified(x_vals)

# Graficar los resultados
plt.figure(figsize=(10, 6))
plt.plot(x_vals, y_vals, 'b-', linewidth=2, label='Serie de Fourier (10 términos)')
plt.title('Aproximación de Onda Cuadrada con Serie de Fourier')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True)
plt.legend()
plt.axhline(y=0, color='k', linestyle='-', alpha=0.3)
plt.axvline(x=0, color='k', linestyle='-', alpha=0.3)
plt.xlim(-2*np.pi, 2*np.pi)
plt.ylim(-1.5, 1.5)
plt.show()
```
